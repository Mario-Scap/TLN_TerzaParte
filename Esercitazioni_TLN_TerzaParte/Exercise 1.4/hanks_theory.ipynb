```python
import collections
from nltk.corpus import brown
from nltk.corpus import wordnet as wn
import spacy
import numpy as np
import matplotlib.pyplot as plt
from nltk import Counter
from collections import Counter
```

```python
nlp = spacy.load("en_core_web_sm")
verb = ["read", "reads", "readen"]
sentences = [sent for sent in brown.sents() if verb[0] in sent or verb[1] or verb[2]]
```

```python
#Lesk Algorithm for WSD
def lesk_algorithm(word, sentence):
    best_sense = []
    max_overlap = -1

    ctx = set(sentence.split())

    for sys in wn.synsets(word):
        signature = set(sys.definition().split())

        for example in sys.examples():
            signature.update(example.split())
        overlap = len(ctx & signature)

        if overlap > max_overlap:
            best_sense = sys
            max_overlap = overlap
    
    return best_sense
```

```python
pairs = [] #subj, dobj supersense
for sent in sentences:
    subj =''
    obj =''

    root = False
    s = ' '.join(sent)

    for token in nlp(s):
        if token.dep_ == 'ROOT' and token.text in verb:
            root = True
            print(sent)
            print()
        elif token.dep_ == 'nsubj' and token.head.text in verb:
            subj = token.text
        elif token.dep_ == 'dobj' and token.head.text in verb:
            obj = token.text
    if subj and obj and root:
        print ("subj :", subj, "\tobj:", obj)
        subj_syns = lesk_algorithm(subj,s)
        obj_syns = lesk_algorithm(obj,s)

        pairs.append((subj_syns.lexname() if subj_syns else 'noun.person', obj_syns.lexname() if obj_syns else 'noun.person'))

        #print(pairs)
```

```python
print("Coppie Semantiche with count:")
for p in set(pairs):
    print([p, pairs.count(p)])
```
